```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import f
```

```python
n = 139
m = 1000
s1_l = 5.722
s1_w = 4.612
s2_l = 6.161
s2_w = 5.055
alpha = 0.05

f_low = f.ppf(alpha/2, n-1, m-1)
f_high = f.ppf(1-alpha/2, n-1, m-1)

f_len = s1_l**2 / s2_l**2
f_wid = s1_w**2 / s2_w**2

print(f"F-статистика ширины: {f_wid}")
print(f"F-статистика длины: {f_len}")
print(f"Gкр: <={f_low} и >={f_high}")
```

$F_{len}$ и $F_{wid}$ $\notin$ Gкр $\implies$ Нет оснований отвергнуть H0

График мощности критерия

```python
plt.figure(figsize=(10, 6))

sigma_ratio = np.linspace(0.5, 2, 100)  

power = 1 - f.cdf(f_high, n-1, m-1, scale=sigma_ratio) + f.cdf(f_low, n-1, m-1, scale=sigma_ratio)

plt.axvline(x=f_len, color='blue', linestyle=':', alpha=0.7, label=f'Длина (x={f_len:.2f})')
plt.axvline(x=f_wid, color='green', linestyle=':', alpha=0.7, label=f'Ширина (x={f_wid:.2f})')

plt.plot(sigma_ratio, power, 'r-', linewidth=2, label='Мощность критерия')

plt.title('График мощности критерия')
plt.xlabel('Отношение дисперсий (σ1²/σ2²)')
plt.ylabel('Мощность критерия')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.01)

plt.show()
```
